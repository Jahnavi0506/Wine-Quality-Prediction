#Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
#Loading dataset in pandas Dataframe
wine_dataset=pd.read_csv('/content/winequality-red.csv')
```

```python
#checking no of rows and cols
wine_dataset.shape
```

```python
correlation=wine_dataset.corr()
```

```python
#Contructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(6,6))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Data Preprocessing

```python
x=wine_dataset.drop('quality',axis=1)
```

```python
print(x)
```

Label Binarization

```python
y=wine_dataset['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
```

```python
print(y)
```

Train & Test Split

```python
#spliting data into training and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

Model Training
#Random Forest Classifier Model

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

Accuracy Score

```python
#Evaluating the model
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy:',test_data_accuracy)
```

Building a predictive system

```python
input_data=(7.3,0.65,0.0,1.2,0.068,15.0,21.0,0.9946,3.39,0.47,10.0)

#changing the input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)
#reshape the data  as we are predicting the label only for  one instaance
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshape)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
